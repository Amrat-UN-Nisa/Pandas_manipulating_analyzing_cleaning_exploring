**Cleaning columns Date of Publication using the .apply function**:

```python
unwanted_characters = ['[', ',', '-','?',']']

def date_cleaning(item):   #item represent the row from the dataframe
    drop= str(item.loc['Date of Publication'])  #access the value and convert it in to string

    if drop == 'nan' or drop[0] == '[':      #drop == 'nan' data is missing or not available drop == '[' check formatting issue
        return np.NaN

    for character in unwanted_characters:
        if character in drop:
            character_index = drop.find(character) #index of the character
            drop = drop[:character_index]    # removes the unwanted characters from the date.

    return drop

df['Date of Publication'] = df.apply(date_cleaning, axis = 1) #axis=1 (rows)
```

```python
#alternate way of cleaning Date of Publication
#run cell to see output
unwanted_characters = ['[', ',', '-','?',']']

def data_cleaning(drop):
    drop = str(drop)
    if drop.startswith('[') or drop == 'nan':
        return 'NaN'
    for character in unwanted_characters:
        if character in drop:
            character_index = drop.find(character)
            drop = drop[:character_index]
    return drop

df['Date of Publication'] = df['Date of Publication'].apply(data_cleaning)
df.head()
```

```python
from functools import reduce
```

**Cleaning and formatting author names.**

```python
def cleaning_author_names(author):
    author = str(author)
    if author == 'nan':
        return 'NaN'
    author = author.split(',')
    if len(author) == 1:
        name = filter(lambda x: x.isalpha(), author[0])     #to filter out alphabetic characters from the name checks if x is an alphabetic character
        return reduce(lambda x, y: x + y, name)             #concatenate

    last_name, first_name = author[0], author[1]

    first_name = first_name[:first_name.find('-')] if '-' in first_name else first_name
    #It's extracting the part of the first name before the hyphen, if a hyphen is present

    if first_name.endswith(('.', '.|')):
        parts = first_name.split('.')

        if len(parts) > 1:    #it extracts the portion of the first_name before the last dot.
            first_occurence = first_name.find('.')
            final_occurence = first_name.find('.', first_occurence + 1)
            first_name = first_name[:final_occurence]
        else:
            first_name = first_name[:first_name.find('.')]

    last_name = last_name.capitalize()

    return f'{first_name} {last_name}'


df['Author'] = df['Author'].apply(cleaning_author_names)
```

```python
def cleaning_title(title):

    if title is None or title == 'nan':
        return 'NaN'

    if title[0] == '[':
        title = title[1: title.find(']')]   # it removes the first character and all characters up to the first occurrence of ']'

    if 'by' in title:
        title = title[:title.find('by')]  # it removes everything from the occurrence of 'by' onward
    elif 'By' in title:
        title = title[:title.find('By')] #it removes everything from the occurrence of 'By' onward

    if '[' in title:
        title = title[:title.find('[')]  #it removes everything from the occurrence of [ onward. This might be used to remove any remaining bracketed content.



    title = title[:-2]   #removes the last two characters from the title

    title = list(map(str.capitalize, title.split())) #This line splits the title into words using spaces as delimiters and then capitalizes the first letter of each word
    return ' '.join(title)

df['Title'] = df['Title'].apply(cleaning_title)
df.head()
```
