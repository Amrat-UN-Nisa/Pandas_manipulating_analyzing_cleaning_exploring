```python
#skiprows=1, the first row of the CSV file will be skipped
#header=0 the first row of the CSV file will be treated as the header.
df  = pd.read_csv("/content/drive/My Drive/DataAnalysisWithPython/Libraries/DataCleaningProcess/DataCleaningProcess01/olympicss.csv",skiprows = 1, header = 0)
```

```python
df.head()
```

**Rename the columns:**

```python
new_names =  {'Unnamed: 0': 'Country',
              '? Summer': 'Summer Olympics',
              '01 !': 'Gold',
              '02 !': 'Silver',
              '03 !': 'Bronze',
              '? Winter': 'Winter Olympics',
              '01 !.1': 'Gold.1',
              '02 !.1': 'Silver.1',
              '03 !.1': 'Bronze.1',
              '? Games': '# Games',
              '01 !.2': 'Gold.2',
              '02 !.2': 'Silver.2',
              '03 !.2': 'Bronze.2'}
df.rename(columns = new_names, inplace = True)
```

```python
df.head()
```

```python
#set_index() set the index of a DataFrame to a column named "Identifier"
df.set_index('Country', inplace = True)
df.head()
```
