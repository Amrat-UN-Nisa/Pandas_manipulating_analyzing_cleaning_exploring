```python
def remove_writby(text):
    return ' '.join(re.sub( r"([A-Z])", r" \1", text[10:]).split())
    #re.sub() search for a pattern in a string
    #re.sub(pattern, replacement, string)
    #pattern: The regular expression pattern you want to search for.
    #replacement: The string that should replace the matched pattern.
    #string: The input string in which you want to perform the substitution
```

```python
def rm_mins(text):
    text.remove('mins') #The remove() method is used to remove a specific element from a list, but it will raise a ValueError if the element is not found in the list.
    return text

def rm_min(text):
    text.remove('min')
    return text

def sixty_hr(text):
    text[text.index('hr')] = '60'
    return text

def sixty_hrs(text):
    text[text.index('hrs')] = '60'
    return text
```

```python
time_list = list()

for i in df['time']:
    i = i.split()
    # Removes 'mins' or 'min' from the text
    if 'mins' in i:
        i = rm_mins(i)
    elif 'min' in i:
        i = rm_min(i)
    # Transforms 'hr' or 'hrs' into int(60)
    if 'hr' in i:
        i = sixty_hr(i)
    elif 'hrs' in i:
        i = sixty_hrs(i)

    try:
        i.remove('and')
        time_list.append(i)
    except ValueError:
        time_list.append(i)
```

```python
output = []

for i in time_list:
    if '60' in i:
        # Find the index of '60' and convert the preceding digit to an integer
        hours = int(i[i.index('60') - 1])
        # Get the last digit and convert it to an integer
        minutes = int(i[-1])
        total_minutes = (hours * 60) + minutes
        output.append(total_minutes)
    else:
      try:
        output.append(int(i[0]))
      except ValueError:
        output.append(i)
print(output)
```

```python
def remove_num(text):
    return re.findall(r'\d+', text)   #\d for digits + means one or more occurance.
```

```python
rating_list = list()
for i in df2['stars']:
    rating_list.append(remove_num(i))
rating_list[0:10]
```

```python
rating = list()
raters = list()
for i in rating_list:
    if len(i) == 4:
        rating.append(float(f'{i[0]}.{i[1]}'))
        raters.append(int(i[3]))
    elif len(i) == 3:
        rating.append(float(i[0]))
        raters.append(int(i[2]))
    else:
        rating.append(0.0)
        raters.append(0)
```

```python
# Assigns the values to a list and then to a series
rated_series = pd.Series(rating)
# Adds a ability 1 col.
df2 = df2.assign(audible_rating = rated_series)
df2.head()
```

```python
# Assigns the values to a list and then to a series
rater_series = pd.Series(raters)
# Adds a ability 1 col.
df2 = df2.assign(rater_number = rater_series)
df2.head()
```
